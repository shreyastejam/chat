# **MultiPOP: Aggregate Method**

```python
arr=[]
n=int(input("Enter the number of elements in array : "))


count=0
def Multipop(n,arr):
    count=0
    for i in range(n):
        arr.pop()
        count+=1
    return count,arr
c1=0
c2=0
for i in range(0,n):
    new=int(input("Enter element : "))
    if i==0:
        arr.append(new)
        c2+=1
        print(arr)
    else:
        if new<=len(arr):
            c,arr=Multipop(new,arr)
            c1+=c
            print("Cost for multipop : ",c)
            print("Array After Multipop : ",arr)
        arr.append(new)
        c2+=1
        print(arr)
print("Total Cost : ",c1+c2)
print("Final Aggregate cost : ",(c1+c2)/n)
```

# Multipop: Accounting method

```python
stack=[]
n=int(input("Enter the number of elements in array: "))
bank=0
assumed_cost=3
cost=0

def Multipop(n,stack):
  count=0
  for i in range(n):
    stack.pop()
    count+=1
  return count,stack


for i in range(0,n):
  bank+=assumed_cost
  new_element=int(input("\nEnter element: "))
  if i==0:
    stack.append(new_element)
    cost+=1
    bank-=1
    print(stack)
    print("Real_Cost=",cost,"\nAssumed_Cost=",assumed_cost,"\nBank=",bank)
  else:
    c=0
    if new_element<=len(stack):
      c,stack=Multipop(new_element,stack)
      cost+=c
      bank-=c
      print("\nStack after Multipop: ",stack)
    stack.append(new_element)
    cost+=1
    bank-=1
    print("Stack=",stack)
    print("Real_Cost=",c+1,"\nAssumed_Cost=",assumed_cost,"\nBank=",bank,"\n")

# print("Total cost: ",cost)
print("\nTotal Real_Cost:",cost,"\nTotal Assumed_Cost:",n*assumed_cost,"\nBank:",bank)
```

**Randomised QuickSort**

```python
import time
import copy
import random
import sys

sys.setrecursionlimit(10**9)

def partition(arr,low,high):
   i = ( low-1 )
   pivot = arr[high]
   for j in range(low , high):
      if arr[j] <= pivot:
         i = i+1
         arr[i],arr[j] = arr[j],arr[i]
   arr[i+1],arr[high] = arr[high],arr[i+1]
   return ( i+1 )


def quickSort(arr,low,high):
   if low < high:
      pi = partition(arr,low,high)
      quickSort(arr, low, pi-1)
      quickSort(arr, pi+1, high)

def partitionrand(arr,low,high):
    randpivot = random.randrange(low, high)
    arr[high], arr[randpivot] = arr[randpivot], arr[high]
    return partition(arr, low, high)

def randquickSort(arr,low,high):
   if low < high:
      pi = partitionrand(arr,low,high)
      quickSort(arr, low, pi-1)
      quickSort(arr, pi+1, high)



# print(arr)
iteration=int(input("Ente number of iterations: "))
for i in range(iteration):
    arr=[]
    n=int(input("\nEnter the length of array: "))
    for i in range(n):
        arr.append(random.randint(1,n))
    arr2=copy.deepcopy(arr)
    # print(arr)
    quick1=time.time()
    quickSort(arr,0,n-1)
    quick2=time.time()-quick1
    print("Time for Quick Sort for ",(i+1)," elements :",quick2)
    # print(arr)
    # print("\n",arr2)
    randquick1=time.time()
    randquickSort(arr2,0,len(arr)-1)
    randquick2=time.time()-randquick1
    print("Time for Randomized Quick Sort for ",(i+1)," elements :",randquick2)
    # print(arr2)
```

```python
# Node creation
class Node():

    def __init__(self, item):
        self.item = item
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1

class RedBlackTree():

    #Initialization
    def __init__(self):
        self.TNULL = Node(0)
        self.TNULL.color = 0
        self.TNULL.left = None
        self.TNULL.right = None
        self.root = self.TNULL

    # Insertion
    def fix_insert(self, k):
        while k.parent.color == 1:
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.left:
                        k = k.parent
                        self.right_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.left_rotate(k.parent.parent)
            else:
                u = k.parent.parent.right

                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.right:
                        k = k.parent
                        self.left_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.right_rotate(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 0

    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.TNULL:
            y.left.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.TNULL:
            y.right.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.item = key
        node.left = self.TNULL
        node.right = self.TNULL
        node.color = 1

        y = None
        x = self.root

        while x != self.TNULL:
            y = x
            if node.item < x.item:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.item < y.item:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.color = 0
            return

        if node.parent.parent == None:
            return

        self.fix_insert(node)

    # Printing
    def __print_helper(self, node, indent, last):
        if node != self.TNULL:
            print(indent, end='')
            if last:
                print("R----", end='')
                indent += "     "
            else:
                print("L----", end='')
                indent += "|    "

            s_color = "RED" if node.color == 1 else "BLACK"
            print(str(node.item) + "(" + s_color + ")")
            self.__print_helper(node.left, indent, False)
            self.__print_helper(node.right, indent, True)

    def print(self):
        self.__print_helper(self.root, "", True)

if __name__ == "__main__":
    bst = RedBlackTree()

    #Initializing
    nodes = [10, 18, 7, 15, 16,30,25,40,60,2,1,70]
    for i in nodes:
        bst.insert(i)
    print("Initial Tree:")
    bst.print()
    print()

    while True:
        print('1. Insert Node')
        print('2. Print Tree')
        print('3. Exit')
        try:
            c = int(input('Enter your choice: ')[0])
        except:
            print('Enter numbers only!')

        if c == 1:
            try:
                n = int(input('Enter the value of node you wish to insert: '))
                bst.insert(n)
            except:
                print('Enter numbers only!')
        elif c == 2:
            bst.print()
        elif c == 3:
            break
        else:
            print('Invalid Input')
        print()
```

**RB Deletion**

```python
# Node creation
class Node():

    def __init__(self, item):
        self.item = item
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1

class RedBlackTree():

    #Initialization
    def __init__(self):
        self.TNULL = Node(0)
        self.TNULL.color = 0
        self.TNULL.left = None
        self.TNULL.right = None
        self.root = self.TNULL

    def fix_insert(self, k):
        while k.parent.color == 1:
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.left:
                        k = k.parent
                        self.right_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.left_rotate(k.parent.parent)
            else:
                u = k.parent.parent.right

                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.right:
                        k = k.parent
                        self.left_rotate(k)
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.right_rotate(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 0


    #Deletion
    def delete_fix(self, x):
        while x != self.root and x.color == 0:
            if x == x.parent.left:
                s = x.parent.right
                if s.color == 1:
                    s.color = 0
                    x.parent.color = 1
                    self.left_rotate(x.parent)
                    s = x.parent.right
                if s.left.color == 0 and s.right.color == 0:
                    s.color = 1
                    x = x.parent
                else:
                    if s.right.color == 0:
                        s.left.color = 0
                        s.color = 1
                        self.right_rotate(s)
                        s = x.parent.right
                    s.color = x.parent.color
                    x.parent.color = 0
                    s.right.color = 0
                    self.left_rotate(x.parent)
                    x = self.root
            else:
                s = x.parent.left
                if s.color == 1:
                    s.color = 0
                    x.parent.color = 1
                    self.right_rotate(x.parent)
                    s = x.parent.left
                if s.right.color == 0 and s.left.color == 0:
                    s.color = 1
                    x = x.parent
                else:
                    if s.left.color == 0:
                        s.right.color = 0
                        s.color = 1
                        self.left_rotate(s)
                        s = x.parent.left
                    s.color = x.parent.color
                    x.parent.color = 0
                    s.left.color = 0
                    self.right_rotate(x.parent)
                    x = self.root
        x.color = 0

    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.item = key
        node.left = self.TNULL
        node.right = self.TNULL
        node.color = 1

        y = None
        x = self.root

        while x != self.TNULL:
            y = x
            if node.item < x.item:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.item < y.item:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.color = 0
            return

        if node.parent.parent == None:
            return

        self.fix_insert(node)

    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.TNULL:
            y.left.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.TNULL:
            y.right.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    
    def __rb_transplant(self, u, v):
        if u.parent == None:
            self.root = v
        elif u == u.parent.left:
            u.parent.left = v
        else:
            u.parent.right = v
        v.parent = u.parent
    
    def minimum(self, node):
        while node.left != self.TNULL:
            node = node.left
        return node

    def delete_node_helper(self, node, key):
        z = self.TNULL
        while node != self.TNULL:
            if node.item == key:
                z = node
            if node.item <= key:
                node = node.right
            else:
                node = node.left
        if z == self.TNULL:
            print("Cannot find key in the tree")
            return
        y = z
        y_original_color = y.color
        if z.left == self.TNULL:
            x = z.right
            self.__rb_transplant(z, z.right)
        elif (z.right == self.TNULL):
            x = z.left
            self.__rb_transplant(z, z.left)
        else:
            y = self.minimum(z.right)
            y_original_color = y.color
            x = y.right
            if y.parent == z:
                x.parent = y
            else:
                self.__rb_transplant(y, y.right)
                y.right = z.right
                y.right.parent = y
            self.__rb_transplant(z, y)
            y.left = z.left
            y.left.parent = y
            y.color = z.color
        if y_original_color == 0:
            self.delete_fix(x)
    
    def delete(self, item):
        self.delete_node_helper(self.root, item)

    # Printing
    def __print_helper(self, node, indent, last):
        if node != self.TNULL:
            print(indent, end='')
            if last:
                print("R----", end='')
                indent += "     "
            else:
                print("L----", end='')
                indent += "|    "
            s_color = "RED" if node.color == 1 else "BLACK"
            print(str(node.item) + "(" + s_color + ")")
            self.__print_helper(node.left, indent, False)
            self.__print_helper(node.right, indent, True)

    def print(self):
        self.__print_helper(self.root, "", True)

if __name__ == "__main__":
    bst = RedBlackTree()

    #Initializing
    nodes = [2, 4, 16, 13, 21]
    for i in nodes:
        bst.insert(i)
    print("Initial Tree:")
    bst.print()
    print()

    while True:
        print('1. Delete Node')
        print('2. Print Tree')
        print('3. Exit')
        try:
            c = int(input('Enter your choice: ')[0])
        except:
            print('Enter numbers only!')
            
        if c == 1:
            try:
                n = int(input('Enter the value of node you wish to delete: '))
                bst.delete(n)
            except:
                print('Enter numbers only!')
        elif c == 2:
            bst.print()
        elif c == 3:
            break
        else:
            print('Invalid Input')
        print()
```

**MAX Flow**

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph(G):
  pos = nx.spring_layout(G)
  nx.draw_networkx(G, pos, with_labels=True, font_weight='bold')
  labels = nx.get_edge_attributes(G, 'weight')
  nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
  plt.show()

def getMinFlow(G, path):
    min_flow = 100
    for i in range(0, len(path)-1):
        temp = G[path[i]][path[i+1]]['weight']
        if (min_flow > temp): 
            min_flow = temp
    return min_flow


G = nx.DiGraph()
G.add_weighted_edges_from([('S', 'A', 10), ('A', 'B', 4), ('B', 'T', 10), ('S', 'C', 10), ('C', 'D', 9), ('D', 'T', 10), ('A', 'C', 2)])
draw_graph(G)

paths = nx.all_simple_paths(G, "S", "T")
max_flow = 0

for path in paths:
    temp = getMinFlow(G, path)
    print("For path ", path, " having min flow ", temp)
    for i in range (0, len(path) - 1):
        G[path[i]][path[i+1]]['weight'] -= temp
    draw_graph(G)
    max_flow += temp

print("Max Flow: ", max_flow)
```
